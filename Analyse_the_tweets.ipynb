#Hashtags

```python
import re
```

```python
def collect_hashtags(text):
  hashtags = re.findall(r"#[a-zA-Z0-9]+", text) #r is raw data
  return ", ".join(hashtags)
```

```python
tweet['hashtags'] = tweet['tweet'].apply(lambda x: collect_hashtags(x))
```

```python
tweet.sample(5)
```

```python
def collect_mentions(text):
  mentions = re.findall(r"@[a-zA-Z0-9]+", text) #r is raw data
  return ", ".join(mentions)
```

```python
tweet['mentions'] = tweet['tweet'].apply(lambda x: collect_mentions(x))
```

```python
tweet.sample(5)
```

```python
def count_hashtags(text):
  hashtags = re.findall(r"#[a-zA-Z0-9]+", text) #r is raw data
  return len(hashtags)
```

```python
tweet['count_hashtag'] = tweet['tweet'].apply(lambda x: count_hashtags(x))
```

```python
tweet.sample(5)
```

```python
count_hashtags = tweet['count_hashtag'].sum()
print(f"The number of hashtags: {count_hashtags}")
```

```python
def count_mentions(text):
  mentions = re.findall(r"@[a-zA-Z0-9]+", text) #r is raw data
  return len(mentions)
```

```python
tweet['count_mention'] = tweet['tweet'].apply(lambda x: count_mentions(x))
```

```python
count_mentions = tweet['count_mention'].sum()
print(f"The number of mentions: {count_mentions}")
```

```python
all_hashtags = tweet['hashtags'].to_string()
hashtags = re.findall(r"#\w+", all_hashtags)
print(hashtags)
```

# count of hashtags

```python
dict= {}
for word in hashtags:
  if word in dict:
    dict[word] += 1
  else:
    dict[word] = 1
print(dict)
dict_new= sorted(dict.items(),key= lambda x:x[1], reverse= True)
dict_new #in descending order
```

```python
top_five_hashtags = dict_new[:5]
```

```python
top_five_hashtags
```

```python
hashtag, count = zip(*top_five_hashtags)
print(hashtag)
print(count)
```

```python
import matplotlib.pyplot as plt
plt.bar(hashtag, count)
plt.show()
```

#sentiment

```python
tweet.head()
```

```python
import textblob
```

```python
from textblob import TextBlob
```

```python
def measure_sentiment(text):
  blob = TextBlob(text)
  return blob.sentiment.polarity #polarity is giving the positive,negative and neutral
```

```python
tweet['sentiments'] = tweet['tweet'].apply(lambda x: measure_sentiment(x))
```

```python
tweet.sample(5)
```

```python
def sentiment_level(score):
  if score > 0:
    return "positive"
  elif score < 0:
    return "negative"
  else:
    return "neutral"
```

```python
tweet['sentiment_level']= tweet['sentiments'].apply(lambda x: sentiment_level(x))
```

```python
tweet.head(5)
```

```python
count = tweet['sentiment_level'].value_counts()
print(count)
```

```python
count.plot(kind="pie", autopct= '%1.1f')
plt.title("Sentiment Analysis")
plt.show()
```
